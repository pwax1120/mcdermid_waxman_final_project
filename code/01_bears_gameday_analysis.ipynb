```python
import import_ipynb  
import utils         
import pandas as pd
```

```python
chicago_crime_all=pd.read_csv("data/chicago_crime_data.csv")
```

```python
#Soldier Field Coordinates
soldier_field = (-87.6190, 41.8623) 

#Preprocess Crime Data
chicago_processed = utils.preprocess_crime_data(
    chicago_crime_all,
    stadium_coords=soldier_field,
    date_col='Date',
    lat_col='Latitude',
    lng_col='Longitude'
)

#Load Bears Game Dates
bears_games_all = utils.load_all_bears_games()

#Filter to home games only
home_games = bears_games_all[bears_games_all["Location"] == "Home"]

#tag game days within crime df
chicago_tagged = utils.tag_game_windows(
    df=chicago_processed,
    games_df=home_games,
    team_name="Bears",
    date_col='date'
)

#Compute Δ Crime by Distance Bands
chicago_delta_by_distance = utils.compute_distance_crime_delta(
    chicago_tagged,
    bins=[0, 1000, 5000, 10000]
)
```

```python
#Stadium-Centered Radius Map and Δ Crime Bar Chart
figmap,figbar=utils.plot_stadium_radii_map_with_bar_chart(
    stadium_coords=soldier_field,            
    delta_df=chicago_delta_by_distance,     
    radii_meters=[1000, 5000, 10000],
    title="Bears Crime Δ by Radius Around Stadium Chicago",
    zoom=9,
    lat_col='Latitude',
    lng_col='Longitude',
    save_path_map="output/figures/chicago_radius_map.png",
    save_path_bar="output/figures/chicago_bears_gameday_distances_bar.png"
)
figmap.show()
figbar.show()
```
