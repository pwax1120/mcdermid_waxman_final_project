```python
import import_ipynb  
import utils         
import pandas as pd
```

```python
chicago_crime_all=pd.read_csv("data/chicago_crime_data.csv")
```

```python
# 1. Define Soldier Field Coordinates (Chicago)
soldier_field = (-87.6190, 41.8623)  # (longitude, latitude)

# 2. Preprocess Crime Data (Chicago dataset)
chicago_processed = utils.preprocess_crime_data(
    chicago_crime_all,
    stadium_coords=soldier_field,
    date_col='Date',
    lat_col='Latitude',
    lng_col='Longitude'
)

# 3. Load Bears Game Dates
bears_games_all = utils.load_all_bears_games()

# Filter to home games only
home_games = bears_games_all[bears_games_all["Location"] == "Home"]

# 4. Tag Game Windows
chicago_tagged = utils.tag_game_windows(
    df=chicago_processed,
    games_df=home_games,
    team_name="Bears",
    date_col='Date'
)

# 5. Compute Δ Crime by Distance Bands
chicago_delta_by_distance = utils.compute_distance_crime_delta(
    chicago_tagged,
    bins=[0, 1000, 5000, 10000]
)
```

```python
# 6. Plot Stadium-Centered Radius Map with Δ Crime Bar Chart
utils.plot_stadium_radii_map_with_bar_chart(
    stadium_coords=soldier_field,
    crime_df=chicago_tagged,                # used only for geo display
    delta_df=chicago_delta_by_distance,     # precomputed Δ crime values
    radii_meters=[1000, 5000, 10000],
    title="Crime Δ by Radius Around Soldier Field",
    zoom=9,
    lat_col='Latitude',
    lng_col='Longitude'
)
```

This shows other interesting things:
- Chicago sees an increase (around 4/5 %) in crime on game days in the city
- The largest percentage increase in crime still happens in the 1km around the stadium going from an average of 2.3 - 3.7 crimes
