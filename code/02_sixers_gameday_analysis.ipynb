```python
import import_ipynb  
import utils         
import pandas as pd
```

```python
years = [2021,2022,2023,2024,2025]
philly_crime_all = utils.load_philly_crime_data(years)
```

```python
#Stadium Coordinates (Wells Fargo Center)
wells_fargo = (-75.1720, 39.9012)

#Preprocess crime data
philly_processed = utils.preprocess_crime_data(
    philly_crime_all,
    stadium_coords=wells_fargo,
    date_col='dispatch_date_time'
)

#load sixers games
sixers_games_all = utils.load_all_sixers_games()

#filter to home games
home_games = sixers_games_all[sixers_games_all["Location"] == "Home"]

#Tag game days in crime df
philly_tagged = utils.tag_game_windows(
    df=philly_processed,
    games_df=home_games,
    team_name="76ers"
)

#Compute Δ Crime by Distance Bands
philly_delta_by_distance = utils.compute_distance_crime_delta(philly_tagged, bins=[0, 1000, 5000, 10000])

#Plot Stadium-Centered Radius Map with Δ Crime Bar Chart
figmap,figbar=utils.plot_stadium_radii_map_with_bar_chart(
    stadium_coords=wells_fargo,
    delta_df=philly_delta_by_distance,
    radii_meters=[1000, 5000, 10000],
    title="Crime Δ by Radius Around Wells Fargo Center",
    zoom=9,
    save_path_bar="output/figures/philly_sixers_gameday_distances_bar.png"
)
figmap.show()
figbar.show()
```

This shows maybe what we would have expected:
- There is no signficant increase or decrease in citywide crime on gamedays
- There is no change in crime around the stadium

Explanations:
- Games are less important
- People care less about the sport and the outcome
- Lower attendance in the stadium
